```python
name = str(input("Name of Employee"))
age = int(input("Enter the employee's age:"))
experience = int(input("Enter the number of years of experience:"))


if age>= 55 and experience>= 25:
                 ATR = 5600000
elif age>= 45 and experience>= 20:
                 ATR = 4480000
elif age>= 35 and experience>= 10:
                 ATR = 1500000
else:
                 ATR = 550000

print(f"Employee: {name} has an ATR of N{ATR}")
```
