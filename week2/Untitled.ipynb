```python
import tkinter as tk
from tkinter import messagebox as msgbox

root=tk.Tk()
root.title("ATR CALCULATOR")
root.geometry("400x400")

def submission():
    try:
        age=int(entry1.get())
        experience=int(entry2.get())
    
        if age>= 55 and experience>= 25:
             ATR = 5600000
        elif age>= 45 and experience>= 20:
             ATR = 4480000
        elif age>= 35 and experience>= 10:
             ATR = 1500000
        else:
             ATR = 550000
        msgbox.showinfo("weldone",F"your atr is {ATR}")
        entry1.delete(0,tk.END)
        entry2.delete(0,tk.END)
    except ValueError:
        msgbox.showerror("wrong","seek help")
label1=tk.Label(root, text="Enter age")
label1.pack()
entry1=tk.Entry(root)
entry1.pack()

label2=tk.Label(root, text="Enter experience")
label2.pack()
entry2=tk.Entry(root)
entry2.pack()

button=tk.Button(root, text="submit", pady=20, bg="black",fg="white", font=("arial",30,"bold"),command=submission)
button.pack()

root.mainloop()
```
