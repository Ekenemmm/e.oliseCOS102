```python
import numpy as np

def solve_equations():
    print("Equation Solver")
    print("1. Quadratic (Ax^2 + Bx + C)")
    print("2. Cubic (Ax^3 + Bx^2 + Cx + D)")
    print("3. Quartic (Ax^4 + Bx^3 + Cx^2 + Dx + E)")
    
    choice = input(" 1, 2, or 3: ")

    if choice == "1":
        a = float(input("Enter A: "))
        b = float(input("Enter B: "))
        c = float(input("Enter C: "))
        coeffs = [a, b, c]
        
    elif choice == "2":
        a = float(input("Enter A: "))
        b = float(input("Enter B: "))
        c = float(input("Enter C: "))
        d = float(input("Enter D: "))
        coeffs = [a, b, c, d]
        
    elif choice == "3":
        a = float(input("Enter A: "))
        b = float(input("Enter B: "))
        c = float(input("Enter C: "))
        d = float(input("Enter D: "))
        e = float(input("Enter E: "))
        coeffs = [a, b, c, d, e]
    
    else:
        print("Invalid choice")
        return

   
    roots = np.roots(coeffs)
    print("\nThe roots are:")
    print(roots)

solve_equations()
```
