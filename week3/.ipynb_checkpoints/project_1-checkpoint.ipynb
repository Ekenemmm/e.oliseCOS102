```python
import tkinter as tk
from tkinter import messagebox as msgbox
import pandas as pd

def verify_employee():
    # get user input 
    search_name = name_entry.get().strip()
    search_dept = dept_entry.get().strip()

    try:
        # put the data
        df = pd.read_csv('GIG-logistics(in).csv')
        
        #remove extra space
        df.columns = df.columns.str.strip()

        # checkif Name && Department match a row 
        match = df[(df['Name'].str.lower() == search_name.lower()) & 
                   (df['Department'].str.lower() == search_dept.lower())]

        if not match.empty:
            # get other members of same depRTMENT and exclude user
            dept_members = df[df['Department'].str.lower() == search_dept.lower()]
            colleagues = dept_members[dept_members['Name'].str.lower() != search_name.lower()]['Name'].tolist()
            
            show_welcome_window(search_name, search_dept, colleagues)
        else:
            #  employee does not exist
            msgbox.showinfo("Verification", " sorry, we could not verify your identity.")
            
    except FileNotFoundError:
        msgbox.showerror("Error", "File not found, ensure 'GIG-logistics.csv' is in the same folder as this notebook.")
    except Exception as e:
        msgbox.showerror("Error", f"unexpected error: {e}")

def show_welcome_window(name, dept, colleagues):
    #  new pop up window
    welcome_win = tk.Toplevel(root)
    welcome_win.title("Employee Verified")
    welcome_win.geometry("400x350")

    tk.Label(welcome_win, text=f"Welcome, {name}!", font=("Arial", 14, "bold"), fg="red").pack(pady=10)
    tk.Label(welcome_win, text=f"Department: {dept}", font=("Arial", 10)).pack()
    
    tk.Label(welcome_win, text="\nOther members in your department:", font=("Arial", 10, "bold")).pack()
    
    # Display colleagues in a Listbox
    listbox = tk.Listbox(welcome_win, width=45, height=10)
    listbox.pack(pady=10)
    
    if colleagues:
        for person in colleagues:
            listbox.insert(tk.END, f" • {person}")
    else:
        listbox.insert(tk.END, "You are the only member in this department.")

#  GUI Setup
root = tk.Tk()
root.title("GIG Logistics - Identity Verification")
root.geometry("350x300")

# header
tk.Label(root, text="GIG LOGISTICS", font=("Helvetica", 16, "bold"), fg="red").pack(pady=15)

# Name Input
tk.Label(root, text="Enter Your Full Name:").pack()
name_entry = tk.Entry(root, width=30)
name_entry.pack(pady=5)

# Dept Input
tk.Label(root, text="Enter Your Department:").pack()
dept_entry = tk.Entry(root, width=30)
dept_entry.pack(pady=5)

# Verify button
verify_btn = tk.Button(root, text="VERIFY", command=verify_employee, bg="black", fg="white", width=15)
verify_btn.pack(pady=20)

root.mainloop()
```
