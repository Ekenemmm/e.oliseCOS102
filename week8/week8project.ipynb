```python
class Zenith:
    def __init__(self, employee_name, division):
        self.employee_name = employee_name
        self.division = division

    def mutual_services(self):
        # common services shared 
        return [
            "Lines of credit", 
            "Investment management and accounts", 
            "Insurance"]

    def unique_services(self):
            #Polymorphism
        pass

# Inheritance

class RetailBanking(Zenith):
    def unique_services(self):
        return [
            "Retirement and education accounts",
            "Loans and mortgages",
            "Checking and saving"     ]

class GlobalBanking(Zenith):
    def mutual_services(self):
        #  we override here
        return ["Liquidity management"]

    def unique_services(self):
        return [
            "Multi-currency management services and products",
            "Foreign currency accounts",
            "Foreign currency credit cards",
            "Transborder advisory services"
        ]

class CommercialBanking(Zenith):
    def unique_services(self):
        return ["advisory services"]

# --- Main Program Logic ---

def run_portal():
    print("Zenith Bank Nigeria Employee Portal ")
    name = input("Enter Employee Name: ")
    print("\ Division:")
    print("1. Retail Banking")
    print("2. Global Banking")
    print("3. Commercial Banking")
    
    choice = input("\nPick choice (1, 2,3): ")

    # object Creation based on input
    if choice == '1':
        employee = RetailBanking(name, "Retail Banking")
    elif choice == '2':
        employee = GlobalBanking(name, "Global Banking")
    elif choice == '3':
        employee = CommercialBanking(name, "commercial Banking")
    else:
        print("Invalid ")
        return

    # dosplay Results

    print(f"Employee Name: {employee.employee_name}")
    print(f"Division:      {employee.division}")
  
    
    print("\nMutual Services rendered:")
    for service in employee.mutual_services():
        print(f" - {service}")

    print("\nUnique Services rendered:")
    for service in employee.unique_services():
        print(f" - {service}")
  

if __name__ == "__main__":
    run_portal()
```
